```python
scene_dir = 'kornia/nerf/kornia_nerf_data/head_sculpture/'
img_ext = 'jpeg'

scene_dir = os.path.normpath(scene_dir)
scene_name = os.path.basename(scene_dir)
print(scene_name)

scene_dir_head_tail = os.path.split(scene_dir)
print(scene_dir_head_tail)

downscale = 4
resampled_scene_name = f'{scene_name}{downscale}'
print(resampled_scene_name)

resampled_scene_dir = os.path.join(scene_dir_head_tail[0], resampled_scene_name)
print(resampled_scene_dir)

img_dir = os.path.join(scene_dir, 'images')
resampled_img_dir = os.path.join(resampled_scene_dir, 'images')
print(resampled_img_dir)
```

```python
img_paths = sorted(glob.glob(f'{img_dir}/*.{img_ext}'))

img0 = read_image(img_paths[0])
print(img0.shape)

if not os.path.exists(resampled_img_dir):
  os.makedirs(resampled_img_dir)

height = int(img0.shape[1] / downscale)
width = int(img0.shape[2] / downscale)
transform = torchvision.transforms.Resize((height, width))

for img_path in img_paths:
    img = read_image(img_path)
    img = transform(img)
    
    resmapled_img_filename = os.path.basename(img_path)
    resmapled_img_path = os.path.join(resampled_img_dir, resmapled_img_filename)

    # print(resmapled_img_path)

    write_jpeg(img, resmapled_img_path)
```
